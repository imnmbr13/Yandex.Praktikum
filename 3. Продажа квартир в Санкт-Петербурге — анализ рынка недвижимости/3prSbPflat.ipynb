## Предобработка данных

### Замена пропусков

Начнем работу с переименования столбцов, `cityCenters_nearest` переименуем в `city_centers_nearest`, `parks_around3000` в `parks_around_3km`, а `ponds_around3000` в `ponds_around_3km`. И проверим работу кода.

```python
data = data.rename(columns={'cityCenters_nearest':'city_centers_nearest', 
                            'parks_around3000':'parks_around_3km', 
                            'ponds_around3000':'ponds_around_3km'}) #передаим словарь старых значений и новых для переименования
print(data.columns) #выведем список названий столбцов
```

В столбцах `floors_total` и  `locality_name` удалим строки с пустыми значениями. Эти данные нам необходимы для дальнейшего исследования, а их объем (0,36% и 0,20%) некритичны для нашего датасета. Восполнить такие данные корректно не получится.

```python
data = data.dropna(subset = ['floors_total', 'locality_name']) #удаляем строки с пустыми значениями в floors_total и locality_name
data.info() #запрашиваем информацию о датасете для контроля
```

Разберемся со значениями Nan в столбце `balcony`. Если бы в квартире был предусмотрен балкон, то пользователь наверняка отметил, ведь значение балкон можно рассмотреть как небольшое преимущество квартиры. Следовательно, там, где есть пустые значения наличия балкона - можем заменить на 0. Затем проверим значения.

```python
data['balcony'] = data['balcony'].fillna(0) #заполним NaN на 0 в столбце balcony
data['balcony'].sort_values().unique() #получим сортированный список значений в столбце balcony
```

Главное отличие апартаментов от квартир — это правовой статус. Юридически апартаменты относятся к нежилой недвижимости, как, например, кафе и офисы. При продаже установка соотвествующего параметра важна, и продавец апартаментов проставил бы его, в тоже время продавцы квартир не обратили бы на него внимание. Проставим в пустых значениях атрибут False.

```python
data['is_apartment'] = data['is_apartment'].fillna(False) #Заменяем Nan на значения False в is_apartment
```

Проверим замену данных, и посмотрим, что стало с типом данных в столбце.

```python
print(data['is_apartment'].unique()) #проверим уникальные значения
print(data['is_apartment'].dtypes) #проверим тип данных
```

Замена произошла корректно и тип данных столбца поменялся с object на bool.

Разберемся с высотой потолков, мы имеем 9195 пропусков и 73 некорректных значения. Почему же некорректных? <a href='http://www.norm-load.ru/SNiP/Data1/1/1909/index.htm'>В пункте 1.1 СНИПа</a> сказано *Высота жилых зданий от пола до потолка не менее 2,5 м, для климатических подрайонов IА, IБ, IГ, IД, IIА - не менее климатических подрайонах от пола до пола - не более 3 м, от пола до потолка - не менее 2,7 м*. 

То есть, значения потолков меньше 2,5м мы должны будем привести либо к минимально допустимому значению, либо к какому-либо усредененному. Проверим среднее и медианное значения в данном столбце.

```python
data['ceiling_height'].describe() #получим медианное, среденее и другие значения в столбце ceiling_height
```

Как можно увидеть, среднее составляет 2,77м, медианное 2,65м. Медианное значение удовлетворяет наше условие больше, поэтому пустые и неправильные значения заменим именно на него.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median()) #пустые значения меняем на медианные
data.loc[data['ceiling_height']<2.5, 'ceiling_height']=data['ceiling_height'].median() #значенимя меньше 2,5м тоже меняем на медианные
```

## Расчёты и добавление результатов в таблицу

Добавим в датафрейм столбец `m_price`, с ценой квадратного метра, для этого разделим цену на общую площадь (дополнительно приведем значения в нем в *Int64*, для простоты восприятия):

```python
data['m_price'] = (data['last_price'] / data['total_area']).round(0) #найдем цену за квад. метр и "откинем" лишние нули
data.head(5) #выведем 5 строк датафрейма для контроля
```

Теперь определим день недели, месяц и год публикации объявления, соответственно получим столбцы `ads_d`, `ads_m`, `ads_y`.

```python
data['ads_d'] = pd.DatetimeIndex(data['first_day_exposition']).weekday  #получим столбец со значением дня недели .dt.day_name()
data['ads_m'] = pd.DatetimeIndex(data['first_day_exposition']).month #получим столбец с числовым значением месяца
data['ads_y'] = pd.DatetimeIndex(data['first_day_exposition']).year #получим столбец со значением года
data.head(5) #выведем 5 строк датафрейма для контроля
```

Сейчас займемся вычислением на каком этаже находится квартира, на *первом*, *последнем* или на *другом*. Это нам будет нужно, для проверки гипотезы о зависимости цены и расположения квартиры. Для такого опередления создадим функцию, которая проверяет этаж квартиры из объявления с общей этажностью дома и возварщает соответствующие значения в столбец `flat_floor`.

```python
def flat_floor_function(row):
    '''Функция принимает значения столбцов floors_total и floor, сранивает их, и возращает
    значение о нахождении квартиры на первом, последнем или другом этаже'''
    if row['floor'] == 1:
        return 'первый'
    if row['floors_total'] == row['floor']:
        return 'последний'
    if row['floors_total'] != row['floor']:
        return 'другой'
data['flat_floor'] = data.apply(flat_floor_function, axis = 1) #применим нашу функцию в столбец flat_floor
```

```python
data['flat_floor'].value_counts() #проверим распределение данных
```

Теперь определим соотношение жилой и общей площади, результаты сохраним в столбец `liv_to_tot` , а  отношение площади кухни к общей в `kit_to_tot`.

```python
data['liv_to_tot'] = (data['living_area']/data['total_area']).round(2) #получим результат liv_to_tot с округлением до 2 знаков
data['kit_to_tot'] = (data['kitchen_area']/data['total_area']).round(2) #получим результат kit_to_tot с округлением до 2 знаков
data.head(10) #выведем 10 строк датафрейма для контроля
```
